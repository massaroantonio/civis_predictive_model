```python
import os
import matplotlib.pyplot as plt
import numpy as np
from datetime import datetime, timedelta
from sklearn import linear_model
from scipy import interpolate
from sklearn.tree import DecisionTreeRegressor
%matplotlib inline
```

```python
ceis=['dasindo.csv','fiave.csv','san_giuliano.csv','sol_de_ise.csv']
```

```python
D=[]
for name in ceis:
    f=open(name,'r')
    for l in f:
        l=l.split(',')
        D.append(datetime.strptime(l[0]+' '+str(int(l[1])-1),'%d/%m/%Y %H'))
    f.close()
dates=np.array(list(np.sort(list(set(D)))))
```

# 2013 Forecasts

```python
start=datetime.strptime('01/01/2013 00','%d/%m/%Y %H')
end=datetime.strptime('02/01/2014 00','%d/%m/%Y %H')
dates_fc=[start]
current=start
for i in range(366*24):
    current=current+timedelta(0,3600)
    dates_fc.append(current)
dates_fc=np.array(dates_fc)
```

```python
sw_fc={}
fcst={}
for i in range(1,13):
    if i<10:
        storagedir='/home/amassaro/civis/civis_data/previsioni/2013_out_short_long_wave/0'+str(i)+'/'
    else:
        storagedir='/home/amassaro/civis/civis_data/previsioni/2013_out_short_long_wave/'+str(i)+'/'
    files=os.listdir(storagedir)
    files=[x for x in files if x.split('_')[1]=='4']
    files.sort()
    for f in files:
        ymd=f.split('_')[2]
        h=f.split('_')[3][:2]
        fwd=int(f.split('_')[4][2:-4])
        dt=datetime.strptime(ymd+h,'%Y%m%d%H')+timedelta(0,fwd*3600)+timedelta(0,3600)
        if dt>=datetime.strptime('2013033100','%Y%m%d%H') and dt<=datetime.strptime('2013082623','%Y%m%d%H'):
            dt+=timedelta(0,3600)
        ff=open(storagedir+f,'r')
        c=0
        for l in ff:
            if c>0:
                sw_fc[dt]=float(l.split(':')[-1][:-1])
            c+=1
        fcst[dt]=fwd
actual_dates_fc=np.array(sw_fc.keys())
actual_dates_fc=np.sort(actual_dates_fc)
```

```python
sw_rad_fc={}
for i in range(len(actual_dates_fc)):
    date=actual_dates_fc[i]
    delta_time=3600*3
    if fcst[date]==3:
        sw_rad_fc[date]=(sw_fc[date]*delta_time)/1000
    else:
        sw_rad_fc[date]=(2*sw_fc[date]*delta_time-sw_fc[actual_dates_fc[i-1]]*delta_time)/1000

rad_fc_keys=np.array(sw_rad_fc.keys())
rad_fc_keys=np.sort(rad_fc_keys)
sw_rad_fc_v=np.array([sw_rad_fc[k] if sw_rad_fc[k]>0 else 0  for k in rad_fc_keys])
```

```python
sw_rad_fc_1={}
for i in range(len(actual_dates_fc)):
    date=actual_dates_fc[i]
    delta_time=3600*3
    sw_rad_fc_1[date]=(sw_fc[date]*delta_time)/1000
    
rad_fc_keys_1=np.array(sw_rad_fc_1.keys())
rad_fc_keys_1=np.sort(rad_fc_keys_1)
sw_rad_fc_v_1=np.array([sw_rad_fc_1[k] for k in rad_fc_keys_1])
```

```python
f=open('../meteotn_storico_irradiazione/san_lorenzo_sun')
i=0
S_ceis={}

start=datetime.strptime('01/01/2013 00','%d/%m/%Y %H')
end=datetime.strptime('31/12/2013 23','%d/%m/%Y %H')
for l in f:
    l=l.split(',') 
    d=datetime.strptime(l[0],'%H:%M:%S %d/%m/%Y')
    if d>=start and d<=end:        
        S_ceis[d]=float(l[1].replace(' ',''))

S_ceis_dates=np.array(S_ceis.keys())
S_ceis_dates=np.sort(S_ceis_dates)

S_ceis_3={}
for k in rad_fc_keys:
    S_ceis_3[k]=0
for k in rad_fc_keys:
    d=S_ceis_dates[np.where((S_ceis_dates>k-timedelta(0,3600*3))&(S_ceis_dates<=k))]
    for dd in d:
        S_ceis_3[k]+=S_ceis[dd]

S_ceis_3_keys=np.sort(np.array(S_ceis_3.keys()))
S_ceis_3_v=np.array([S_ceis_3[k] for k in S_ceis_3_keys])
```
